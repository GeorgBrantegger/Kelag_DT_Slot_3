```python
import numpy as np
import pandas as pd
import time
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv("2015_08_24 18.00 M1 SS100%.csv",sep=";",header=7,index_col=False)
raw_data.replace({',': '.'}, regex=True,inplace=True)
time_format = "%m/%d/%Y %I:%M:%S.%f %p"

col_names = ['timestamp','M1-LA','M2-LA','Druck']
df = pd.DataFrame(columns=col_names)

df['timestamp'] = pd.to_datetime(raw_data["Date/Time"],format=time_format).astype(np.int64)/10**9
df['M1-LA'] = pd.to_numeric(raw_data['M1-LA'])/100
df['M2-LA'] = pd.to_numeric(raw_data['M2-LA'])/100
df['Druck'] = pd.to_numeric(raw_data['P-DRL'])
```

```python
t_vec   = np.array(df['timestamp']-df['timestamp'][0])
LA1_vec = np.array(df['M1-LA']) 
LA2_vec = np.array(df['M2-LA'])
p_vec   = np.array(df['Druck']) 
```

```python
%matplotlib widget

plt.plot(t_vec,LA1_vec)
plt.plot(t_vec,LA2_vec)
```
