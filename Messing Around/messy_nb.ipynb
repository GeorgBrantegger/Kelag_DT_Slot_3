```python
import numpy as np
import scipy
from scipy.interpolate import interp2d
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1,1,10)
y = np.linspace(-1,1,10)

XX,YY = np.meshgrid(x,y)

Z = np.cos((XX ** 2 + YY ** 2)*5)
```

```python
%matplotlib widget
fig1 = plt.figure()
ax1 = plt.axes(projection='3d')

ax1.plot_surface(XX, YY, Z,cmap='viridis', edgecolor='none')
```

```python
n = 5

fig2 = plt.figure()
ax2 = plt.axes(projection='3d')

ax2.scatter(XX[n,:], YY[n,:], Z[n,:],cmap='viridis', edgecolor='none')
```

```python
interpol_fun = interp2d(x,y,Z,kind='quintic')

x2 = np.linspace(-1,1,100)
y2 = np.linspace(-1,1,100)

XX2,YY2 = np.meshgrid(x2,y2)


Z2 = np.reshape(interpol_fun(x2,y2),[100,100])
```

```python
fig3 = plt.figure()
ax3 = plt.axes(projection='3d')

ax3.plot_surface(XX2, YY2, Z2,cmap='viridis', edgecolor='none')
```

```python
fig4 = plt.figure()
ax4 = plt.axes(projection='3d')

ax4.scatter(XX2[10*n,:], YY2[10*n,:], Z2[10*n,:],cmap='viridis', edgecolor='none')
```
