```python
import pandas as pd
import numpy as np
from numpy import cos,exp,sin


time_vec = np.linspace(0,100,1000)
h_vec = cos(time_vec)*exp(-0.03*time_vec)

time_df = pd.DataFrame(time_vec,columns=['time'])
h_df = pd.DataFrame(h_vec,columns=['height'])

concat_df = pd.concat([time_df,h_df],axis=1)
```

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=2, cols=1)

fig.add_trace(go.Scatter(x=concat_df['time'],y=concat_df['height']),
              row=1, col=1)
fig.add_trace(go.Scatter(x=concat_df['time'],y=-1*np.ones_like(concat_df['time'])),
              row=2, col=1)
fig.add_trace(go.Scatter(x=concat_df['time'],y=+1*np.ones_like(concat_df['time'])),
              row=2, col=1)

fig.update_xaxes(title_text="xaxis 1 title", row=1, col=1)
fig.update_xaxes(title_text="xaxis 2 title", row=2, col=1)
fig.update_yaxes(title_text="yaxis 1 title", row=1, col=1)
fig.update_yaxes(title_text="yaxis 2 title", row=2, col=1)


fig.show()
```
